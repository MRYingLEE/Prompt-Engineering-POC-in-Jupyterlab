````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
df = pd.read_csv("chips.csv")
df = df.dropna()
df["year"] = df["year"].astype(int)
X = df["year"].values.reshape(-1,1)
y = df["transistors_million"].values.reshape(-1,1)
# Fit a linear regression model using the log of the y data
model = LinearRegression()
model.fit(X, np.log(y))
# Plot the data points and the best fit line
plt.scatter(X, y, marker="x", color="r")
plt.plot(X, np.exp(model.predict(X)), color="b")
plt.yscale("log")
plt.xlabel("Year")
plt.ylabel("Transistors (millions)")
plt.title("Transistor count over time")
plt.show()
````
